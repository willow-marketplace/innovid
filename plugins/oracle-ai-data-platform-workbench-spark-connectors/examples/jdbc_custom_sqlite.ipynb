# `aidp-jdbc-custom` live test — generic `format('jdbc')` (SQLite memory DB)
Smoke test using SQLite (no external infra needed; in-memory DB). Validates the
Spark JDBC plumbing on the cluster — useful before swapping in a customer DB.


```python
import sys, os
# Adjust this if you've uploaded the plugin scripts/ dir elsewhere.
sys.path.insert(0, '/Workspace/Shared/oracle_ai_data_platform_connectors/scripts')
```

```python
JDBC_URL  = 'jdbc:sqlite::memory:'
DRIVER    = 'org.sqlite.JDBC'
props = {'driver': DRIVER, 'user': '', 'password': '', 'fetchsize': '1000'}
df = (spark.read.format('jdbc').options(**props)
      .option('url', JDBC_URL)
      .option('dbtable', '(SELECT 1 c1, 2 c2)').load())
df.show()
```

```python
# Live-test driver parses this final cell's stdout for the JSON summary.
import json, time
summary = {
    'connector': 'aidp-jdbc-custom',
    'auth': 'sqlite-memory',
    'rows': int(df.count()),
    'schema': sorted([f.name for f in df.schema.fields]),
    'timestamp_utc': int(time.time()),
}
print('AIDP_LIVE_TEST_RESULT_BEGIN')
print(json.dumps(summary, indent=2))
print('AIDP_LIVE_TEST_RESULT_END')
```
