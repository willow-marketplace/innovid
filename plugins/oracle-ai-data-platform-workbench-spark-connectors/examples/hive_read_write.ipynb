# Hive Connector Samples

Read/write ingestion samples for the `HIVE` connector.


## Ingestion Sample

Replace placeholders such as `<HOST>`, `<USERNAME>`, `<PASSWORD>`, `<SCHEMA>`, and `<TABLE_NAME>` with values from your environment before running the notebook.


```python
# Read from Hive
hive_df = spark.read.format("aidataplatform") \
    .option("type", "HIVE") \
    .option("host", "<HOST>") \
    .option("port", "<PORT>") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TABLE_NAME>") \
    .load()

hive_df.show()
```

## Write Sample

The write sample uses `CREATE`; switch `write.mode` to `APPEND` or `OVERWRITE` when that matches the target behavior.


```python
# Write to Hive
hive_df.write.format("aidataplatform") \
    .option("type", "HIVE") \
    .option("host", "<HOST>") \
    .option("port", "<PORT>") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TARGET_TABLE_NAME>") \
    .option("write.mode", "CREATE") \
    .save()
```

## Pushdown SQL Sample

Use `pushdown.sql` when you want the connector to execute a complete source SQL query instead of building the query from `schema` and `table` options.


```python
# Full SQL pushdown for Hive
hive_df_pushdown = spark.read.format("aidataplatform") \
    .option("type", "HIVE") \
    .option("host", "<HOST>") \
    .option("port", "<PORT>") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("pushdown.sql", "SELECT * FROM <SCHEMA>.<TABLE_NAME> WHERE <COLUMN_NAME> = '<VALUE>'") \
    .load()

hive_df_pushdown.show()
```

## Connector Options

| Parameter name | Valid values | Mandatory | Description |
| --- | --- | --- | --- |
| `type` | `HIVE` | Y | Data source type. |
| `host` | Valid hostname or IP | Y | Hive host. |
| `port` | Valid port number | Y | Hive port. |
| `database.name` | Database name | N | Optional database name when needed. |
| `authentication.method` | `basic`, `kerberos` | N | Authentication method. If omitted, the connector attempts a non-explicit auth flow. |
| `user.name` | Valid username | Y for basic, N for some Kerberos flows | Username used for connectivity. |
| `password` | Valid password | Y for basic, N for some Kerberos flows | Password used for connectivity. |
| `service.principal.name` | Kerberos principal name | Y for Kerberos | Service principal name for Kerberos authentication. |
| `key.tab.content` | Keytab content | Y for Kerberos | Keytab content used for Kerberos authentication. |
| `key.distribution.center` | Valid hostname or IP | Y for Kerberos | KDC host used for Kerberos authentication. |
| `schema` | Valid schema name | Y | Schema to read from or write to. |
| `table` | Valid table name | Y for table access | Table name inside the schema. |
| `view` | Valid view name | N | View name inside the schema. |
| `sql` | Valid SQL query | N | Custom SQL query for reads. |
| `fetch.size` | Positive integer | N | Number of records fetched per round trip for reads. |
| `row.limit` | Positive integer | N | Limits the number of rows read. |
| `partition.column` | Column name | N | Column used for range-based partitioned reads or writes. |
| `partition.num` | Integer >= 1 | N | Number of partitions to use. |
| `partition.lower` | Numeric value or ISO-8601 date/time | N | Lower bound for range partitioning. |
| `partition.upper` | Numeric value or ISO-8601 date/time | N | Upper bound for range partitioning. |
| `pushdown.columns` | Comma-separated column names | N | Restricts the read to only the requested columns. |
| `pushdown.filter` | SQL predicate | N | Predicate pushed down to the source. |
| `pushdown.sql` | SQL query | N | Full SQL pushdown query. |
| `write.mode` | `CREATE`, `APPEND`, `OVERWRITE` | Y during write | Write mode to use. |
| `write.empty.value.as.null` | `true`, `false` | N | Convert empty incoming string values to null before writing. |
| `write.batch.size` | Positive integer | N | Number of rows written in a batch. |
| `write.reject.limit` | Non-negative integer | N | Maximum rejected-row threshold before the write fails. |
| `login.timeout.seconds` | Integer from `1` to `300` | N | JDBC login timeout in seconds. |
