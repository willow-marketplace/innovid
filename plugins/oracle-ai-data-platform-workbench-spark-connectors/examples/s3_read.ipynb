# `aidp-aws-s3` live test — `s3a://` via access key
Requires `aws-java-sdk-bundle-<ver>.jar` matching the cluster's `hadoop-aws`.


```python
import sys, os
# Adjust this if you've uploaded the plugin scripts/ dir elsewhere.
sys.path.insert(0, '/Workspace/Shared/oracle_ai_data_platform_connectors/scripts')
```

```python
spark.conf.set('fs.s3a.aws.credentials.provider', 'org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider')
os.environ['AWS_ACCESS_KEY_ID']     = os.environ['S3_ACCESS_KEY']
os.environ['AWS_SECRET_ACCESS_KEY'] = os.environ['S3_SECRET_KEY']
bucket = os.environ['S3_BUCKET']
key    = os.environ['S3_FILE']
df = spark.read.json(f's3a://{bucket}/{key}')
df.show()
```

```python
# Live-test driver parses this final cell's stdout for the JSON summary.
import json, time
summary = {
    'connector': 'aidp-aws-s3',
    'auth': 'access-key',
    'rows': int(df.count()),
    'schema': sorted([f.name for f in df.schema.fields]),
    'timestamp_utc': int(time.time()),
}
print('AIDP_LIVE_TEST_RESULT_BEGIN')
print(json.dumps(summary, indent=2))
print('AIDP_LIVE_TEST_RESULT_END')
```
