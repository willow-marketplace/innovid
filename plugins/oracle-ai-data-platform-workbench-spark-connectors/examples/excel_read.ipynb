# `aidp-excel` live test — `pandas → CSV → Spark` (no jars)
Reads an .xlsx via pandas (driver-side), converts to CSV in the same path, then
re-reads as Spark. The `com.crealytics.spark.excel` path is also documented in
the SKILL but requires a cluster JAR install.


```python
import sys, os
# Adjust this if you've uploaded the plugin scripts/ dir elsewhere.
sys.path.insert(0, '/Workspace/Shared/oracle_ai_data_platform_connectors/scripts')
```

```python
import pandas as pd
excel_path = os.environ['EXCEL_PATH']
csv_path   = excel_path.replace('.xlsx', '.csv')
pdf = pd.read_excel(excel_path)
pdf.to_csv(csv_path, index=False)
print('rows in pandas:', len(pdf))
df = spark.read.csv(csv_path, header=True, inferSchema=True)
df.show()
```

```python
# Live-test driver parses this final cell's stdout for the JSON summary.
import json, time
summary = {
    'connector': 'aidp-excel',
    'auth': 'pandas-csv-fallback',
    'rows': int(df.count()),
    'schema': sorted([f.name for f in df.schema.fields]),
    'timestamp_utc': int(time.time()),
}
print('AIDP_LIVE_TEST_RESULT_BEGIN')
print(json.dumps(summary, indent=2))
print('AIDP_LIVE_TEST_RESULT_END')
```
