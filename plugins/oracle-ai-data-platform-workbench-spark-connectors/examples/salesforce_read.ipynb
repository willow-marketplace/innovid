# Salesforce Connector Samples

Read-only ingestion samples for the `SFORCE` connector.


## Ingestion Sample

Replace placeholders such as `<HOST>`, `<USERNAME>`, `<PASSWORD>`, `<SCHEMA>`, and `<TABLE_NAME>` with values from your environment before running the notebook.


```python
# Read from Salesforce
salesforce_df = spark.read.format("aidataplatform") \
    .option("type", "SFORCE") \
    .option("host", "<HOST>") \
    .option("port", "<PORT>") \
    .option("database.name", "<DATABASE_NAME>") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("schema", "<SCHEMA>") \
    .option("table", "<TABLE_NAME>") \
    .load()

salesforce_df.show()
```

## Pushdown SQL Sample

Use `pushdown.sql` when you want the connector to execute a complete source SQL query instead of building the query from `schema` and `table` options.


```python
# Full SQL pushdown for Salesforce
salesforce_df_pushdown = spark.read.format("aidataplatform") \
    .option("type", "SFORCE") \
    .option("host", "<HOST>") \
    .option("port", "<PORT>") \
    .option("database.name", "<DATABASE_NAME>") \
    .option("user.name", "<USERNAME>") \
    .option("password", "<PASSWORD>") \
    .option("pushdown.sql", "SELECT * FROM <SCHEMA>.<TABLE_NAME> WHERE <COLUMN_NAME> = '<VALUE>'") \
    .load()

salesforce_df_pushdown.show()
```

## Connector Options

| Parameter name | Valid values | Mandatory | Description |
| --- | --- | --- | --- |
| `type` | `SFORCE` | Y | Data source type. |
| `host` | Valid Salesforce JDBC host or endpoint | Y | Salesforce endpoint host. |
| `port` | Valid port number | Y | Network port used by the Salesforce JDBC endpoint. |
| `database.name` | Optional Salesforce database or catalog name exposed by the driver | N | Optional logical database name when the Salesforce JDBC configuration expects it. |
| `user.name` | Valid Salesforce username | Y | Username used for Salesforce connectivity. |
| `password` | Valid Salesforce password or security-token-appended password, depending on the configured endpoint | Y | Password used for Salesforce connectivity. |
| `schema` | Valid schema name | Y | Schema exposed by the Salesforce JDBC driver. |
| `table` | Valid object or table name | Y for table access | Salesforce object name to read. |
| `view` | Valid view name | N | View name when the driver exposes the target object as a view. |
| `sql` | Valid SQL query | N | Custom SQL query for reads. |
| `fetch.size` | Positive integer | N | Number of records fetched per round trip for reads. |
| `row.limit` | Positive integer | N | Limits the number of rows read. |
| `partition.column` | Column name | N | Column used for range-based partitioned reads. |
| `partition.num` | Integer >= 1 | N | Number of partitions to use for reads. |
| `partition.lower` | Numeric value or ISO-8601 date/time | N | Lower bound for range partitioning. |
| `partition.upper` | Numeric value or ISO-8601 date/time | N | Upper bound for range partitioning. |
| `pushdown.columns` | Comma-separated column names | N | Restricts the read to only the requested columns. |
| `pushdown.filter` | SQL predicate | N | Predicate pushed down to the source. |
| `pushdown.sql` | SQL query | N | Full SQL pushdown query. |
| `login.timeout.seconds` | Integer from `1` to `300` | N | JDBC login timeout in seconds. |
